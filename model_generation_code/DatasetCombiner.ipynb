```python
# Generamos un diccionario asignando un label en funci√≥n del subreddit
label = {
    'ptsd': 7,
    'EDAnonymous': 6,
    'adhd': 1,
    'depression': 5,
    'schizophrenia': 8,
    'bpd': 4,
    'bipolarreddit': 3,
    'anxiety': 2
}
```

```python
import os
import pandas as pd

folder_path = '/content/drive/Shareddrives/DL + NLP/Proyecto DL + NLP/dataset/RedditMHD/Data'

csv_files = []
for dirpath, dirnames, filenames in os.walk(folder_path):
    for filename in [f for f in filenames if f.endswith('.csv')]:
        csv_files.append(os.path.join(dirpath, filename))
```

```python
df_list = []
for csv_file in csv_files:
    df = pd.read_csv(csv_file, usecols=['post', 'subreddit'])
    df_list.append(df)

combined_df = pd.concat(df_list)
```

```python
combined_df['label'] = combined_df['subreddit'].map(label).fillna(0).astype(int)
```

```python
combined_df = combined_df[['subreddit', 'post', 'label']]
```

```python
combined_df['subreddit'].unique()
```

```python
combined_df['subreddit'] = combined_df['subreddit'].str.replace('parenting','control')
combined_df['subreddit'] = combined_df['subreddit'].str.replace('teaching','control')
combined_df['subreddit'] = combined_df['subreddit'].str.replace('relationships','control')
combined_df['subreddit'] = combined_df['subreddit'].str.replace('personalfinance','control')
combined_df['subreddit'] = combined_df['subreddit'].str.replace('meditation','control')
combined_df['subreddit'] = combined_df['subreddit'].str.replace('legaladvice','control')
combined_df['subreddit'] = combined_df['subreddit'].str.replace('fitness','control')
```

```python
combined_df['subreddit'].unique()
```

```python
combined_df['subreddit'].value_counts()
```

```python
from sklearn.utils import shuffle
combined_df = shuffle(combined_df)
```

```python
control_df = combined_df.loc[combined_df['subreddit'] == 'control']
depression_df = combined_df.loc[combined_df['subreddit'] == 'depression']
anxiety_df = combined_df.loc[combined_df['subreddit'] == 'anxiety']
adhd_df = combined_df.loc[combined_df['subreddit'] == 'adhd']
bpd_df = combined_df.loc[combined_df['subreddit'] == 'bpd']
EDAnonymous_df = combined_df.loc[combined_df['subreddit'] == 'EDAnonymous']
schizophrenia_df = combined_df.loc[combined_df['subreddit'] == 'schizophrenia']
ptsd_df = combined_df.loc[combined_df['subreddit'] == 'ptsd']
bipolarreddit_df = combined_df.loc[combined_df['subreddit'] == 'bipolarreddit']
```

```python
control_df = control_df.head(16000)
depression_df = depression_df.head(16000)
anxiety_df = anxiety_df.head(16000)
adhd_df = adhd_df.head(16000)
bpd_df = bpd_df.head(16000)
EDAnonymous_df = EDAnonymous_df.head(16000)
schizophrenia_df = schizophrenia_df.head(16000)
ptsd_df = ptsd_df.head(16000)
bipolarreddit_df = bipolarreddit_df.head(16000)
```

```python
df_list = [control_df, depression_df, anxiety_df, adhd_df, bpd_df, EDAnonymous_df, schizophrenia_df, ptsd_df, bipolarreddit_df]
```

```python
combined_filtered_df = pd.concat(df_list)
```
