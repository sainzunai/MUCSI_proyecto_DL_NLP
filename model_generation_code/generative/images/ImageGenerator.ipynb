Pipeline

```python
import torch
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", torch_dtype=torch.float16)  
```

```python
pipe = pipe.to("cuda")
```

```python
#Cuidado
pipe.safety_checker = lambda images, clip_input: (images, False)
```

Generación de imágenes

```python
import os

```

```python
topics = ["Schizophrenia", "ADHD", "Anxiety", "Depression"]

for topic in topics:

  os.mkdir(str(topic))
```

```python
topics = ["Schizophrenia", "ADHD", "Anxiety", "Depression"]

num_images = 15

seed = 100

for i in range(len(topics)):
  topic = topics[i]

  if topic == "Schizophrenia":
    prompt = "Schizophrenia as a concepty"

  if topic == "ADHD":
    prompt = "ADHD as a concept"

  if topic == "Anxiety":
    prompt = "Anxiety as a concept"

  if topic == "Depression":
    prompt = "Depression as a concept"


  for n in range(num_images):
    generator = torch.Generator("cuda").manual_seed(seed)
    seed += 4
    image = pipe(prompt, num_inference_steps=50, generator=generator).images[0]
    image.save(topic +"/" + str(n) + ".png")
```

Grid de imágenes

```python
from PIL import Image

def image_grid(imgs, rows, cols):
    assert len(imgs) == rows*cols

    w, h = imgs[0].size
    grid = Image.new('RGB', size=(cols*w, rows*h))
    grid_w, grid_h = grid.size
    
    for i, img in enumerate(imgs):
        grid.paste(img, box=(i%cols*w, i//cols*h))
    return grid
```

```python
num_images = 3
prompt = ["A person with Schizophrenia"] * num_images

images = pipe(prompt).images

grid = image_grid(images, rows=1, cols=3)
grid
```

```python
num_cols = 3
num_rows = 4

Topic = "Schizophrenia"

prompt = ["A person with " + Topic] * num_cols

all_images = []
for i in range(num_rows):
  images = pipe(prompt).images
  all_images.extend(images)
  image.save(Topic + i + ".png")

grid = image_grid(all_images, rows=num_rows, cols=num_cols)
grid
```
